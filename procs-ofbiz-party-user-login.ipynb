```python
import sagas.ofbiz.entities as ee

params=oc.j.HashMap()
params.put('login.username', 'demo.person')
params.put('login.password', 'ofbiz')
result = oc.dispatcher.runSync('userLogin', params)
ok = oc.j.ServiceUtil.isSuccess(result)
if not ok:
    err = oc.j.ServiceUtil.getErrorMessage(result)
    print(err)
else:
    ee.format(result['userLogin'])
```
