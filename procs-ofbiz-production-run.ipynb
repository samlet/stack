```python
class Val(object):
    def __init__(self):
        pass
    def __getattribute__(self, attr):
        """__getattribute__ is similar to __getattr__, 
        with the important difference that __getattribute__ will 
        intercept EVERY attribute lookup, 
        doesnâ€™t matter if the attribute exists or not."""
        print('->', attr)
        __dict__ = super(Val, self).__getattribute__('__dict__')
        if attr in __dict__:
            return super(Val, self).__getattribute__(attr)        
        return attr
    
serviceCtx=Val()
serviceCtx.productId='PROD_MANUF'
print(serviceCtx.productId)
print(serviceCtx.__dict__)
```

```python
import json
def createProductionRun(**kwargs):
    ctx=oc.jmap(**kwargs)
    json_pars=oc.j.ValueHelper.mapToJson(ctx)

    srv=oc.j.ServiceInvoker(oc.dispatcher, oc.delegator, "createProductionRun", json_pars)
    ret=srv.invoke()
    print(ret, ret.getID())
    json_r=srv.getJsonResult()
    print(json.dumps(json.loads(json_r), indent=2))
```
