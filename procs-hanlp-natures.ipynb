```python
from sagas.bots.hanlp_procs import Hanlp, hanlp, hanlp_c
from py4j.java_gateway import JavaGateway, JavaObject, GatewayParameters
for info in hanlp.helper.getHanlpInfo():
    print(info)

def jlist(j, *args):
    m = j.ArrayList()
    for e in args:
        m.append(e)
    return m
def jmap(j, **kwargs) -> JavaObject:
    m = j.HashMap()
    for key, value in kwargs.items():
        m[key] = value
    return m
def jset(j, *args):
    s = j.HashSet()
    for e in args:
        s.add(e)
    return s
```
