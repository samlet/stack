```python
def lines(filename):
    with open(filename) as f:
        lines = f.readlines()        
        return [line.split('\t') for line in lines]
    
dataf="/pi/ai/seq2seq/jpn-eng/jpn.txt"
pairs=lines(dataf)
```

```python
import sagas.graph.dgraph_helper as helper
import pydgraph
client=helper.reset('sents: string @index(term) .')
```

```python
def filter_term(s):
    return s.replace('"',"'").strip()

def add_term(index, pair, rs, lang='ja'):
    rs.append('<_:s%d> <sents> "%s" .'%(index, filter_term(pair[0])))
    rs.append('<_:s%d> <ja> "%s"@%s .'%(index, filter_term(pair[1]), lang))

# client=helper.reset('sents: string @index(fulltext) .')
client=helper.reset('''
sents: string @index(fulltext) .
ja: string @index(fulltext) @lang .
''')
rs=[]
for i in range(2000,2100):
    add_term(i, pairs[i], rs)
cnt='\n'.join(rs)
# print(cnt)
_=helper.set_nquads(client, cnt)
```

```python
helper.run_q(client, '''{
  data(func: alloftext(sents, "diet")) {
     sents
     ja
  }
}''')
```

```python
helper.run_q(client, '''{
  data(func: alloftext(sents, "on a diet")) {
     sents
     ja
  }
}''')
```

```python
def query_with_vars(client, query, vars):
    response = client.txn().query(query, variables=vars)
    helper.print_rs(response)

vars = {'$a': 'on a diet'}
query_with_vars(client, '''query data($a: string){
  data(func: alloftext(sents, $a)) {
     sents
     ja
  }
}''', vars)
```
