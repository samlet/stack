```python
from sagas.ofbiz.resources import ResourceDigester
rd=ResourceDigester()
props=rd.get_all_properties()
rd.print_property(props['CommonStatus'])
```

```python
print(len(props))
```

```python
def property_json(prop_key, prop):
    rs={'key':prop_key}
    for key in prop.values.keys():
        if key not in ['zh-TW']:
            parts=key.split('-')
            rs['value@'+parts[0]]=prop.values[key]
    return rs
def properties_json(props):
    rs=[]
    for k,v in props.items():
        rs.append(property_json(k,v))
    return rs

import json
jval=json.dumps(property_json('CommonStatus', props['CommonStatus']), 
           indent=2, ensure_ascii=False)
print(jval)
```

```python
import json_utils
json_utils.write_json_to_file(
    './data/labels/labels.json', properties_json(props))
```
