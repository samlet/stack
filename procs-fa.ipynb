```python
def traviz(chunk, lang):
    import sagas
    v=lambda s: sagas.dia(lang, local_translit=True).ana(s)

    sents=chunk.split('\n')[1]
    return v(sents)

traviz('''我们 还 需要 什么 ?
‫دیگر چی لازم داریم؟‬
mâ che chizi digari lâzem dârim?
''', 'fa')
```
