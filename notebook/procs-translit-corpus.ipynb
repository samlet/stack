```python
def traviz(chunk, lang):
    import sagas
    v=lambda s: sagas.dia(lang, local_translit=True).ana(s)

    sents=chunk.split('\n')[1]
    return v(sents)

traviz(chunk='''我们 学习 一门 语言 。
‫ہم ایک زبان سیکھ رہے ہیں-‬
hum aik zabaan seekh rahay hin-
''', lang='ur')
```

```python
from sagas.nlu.transliterations import translits
print(translits.trans_polyglot('‫ہم کیا کررہے ہیں؟‬', 'ur'))
print(translits.trans_icu('‫ہم کیا کررہے ہیں؟‬'))
```
