```python
from sagas.ofbiz.services import OfService as s, oc, track
from sagas.ofbiz.entities import OfEntity as e
from py4j.java_gateway import get_field

oc.import_package('org.apache.ofbiz.base.component.ComponentConfig')
allComponents=oc.j.ComponentConfig.getAllComponents()
index=1
comp_files={}
for c in allComponents:
    comp_files[c.getGlobalName()]=[]
for c in allComponents:
    # print(index, c.getComponentName(), c.getRootLocation())
    print(index, c.getGlobalName(), c.getRootLocation())
    index=index+1
    ts=c.getTestSuiteInfos()
    print('\t', 'contains test suites', len(ts))
    ent_res=c.getEntityResourceInfos()
    for es in ent_res:
        data_type=get_field(es,'type')
        if data_type=='data':
            print('\t', data_type, get_field(es,'readerName'),
                 es.getLocation())
            comp_files[c.getGlobalName()].append(c.getRootLocation()+es.getLocation())
```

```python
import os
import xml.etree.ElementTree as ET
import json

def get_data_root(component, group_name):
    files=comp_files[component]
    for f in files:
        base=os.path.basename(f)
        name=os.path.splitext(base)[0]
        if name==group_name:
            # print(name, f)    
            tree = ET.parse(f)
            root = tree.getroot()
            return root
    return None

# root=get_data_root('ofbizDemo', 'OfbizDemoDemoData')
root=get_data_root('ecommerce', 'DemoFactoids')
print(root.tag)
index=0
for child in root:
    # print(child.tag, child.attrib)
    print(index, child.tag)
    for k,v in child.attrib.items():
        print('\t', k, '=', v)
    for cc in child:
        print('\t--', cc.tag, cc.text)
    # print(json.dumps(child.attrib))
    index=index+1
```
