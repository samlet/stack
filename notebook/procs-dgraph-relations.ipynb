```python
from sagas.ofbiz.entities import OfEntity as e, oc
import sagas.ofbiz.entities as ee
import json
import sagas.ofbiz.entity_graph_relations as egr
import json_utils

def proc_rels(entity, key, rel):
    ent = ee.entity(entity)
    val = ent.record(key)
    rs = egr.rel_recs(val, rel, False)
    print('total related recs:', len(rs))
    
def proc_entity(entity, key, rels):
    ent = ee.entity(entity)
    val = ent.record(key)
    ent_r=ent.to_json(val,True, True)
    for rel in rels:
        rs=egr.rel_recs(val, rel, False)
        if len(rs)>0:
            ent_r[rel]=rs
    return ent_r

def proc_entity_recs(entity):
    ent = ee.entity(entity)
    vals=ent.all()
    result=[]
    for val in vals:
        ent_r=ent.to_json(val,True, True)
        rels=ent.get_rel_ent_names(False)
        for rel in rels:
            if egr.is_rel_one(val, rel):
                rv=val.getRelatedOne(rel, False)
                if rv is not None:
                    ent_r[rel]=egr.to_json(rv)
            else:
                rs=egr.rel_recs(val, rel, False)
                if len(rs)>0:
                    ent_r[rel]=rs
        result.append(ent_r)
    return result

proc_rels('Product', 'GZ-2002', 'ProductKeyword')
rs=proc_entity('Product', 'GZ-2002', ['ProductKeyword'])
print(json.dumps(rs, indent=2))

# rs=proc_entity_recs('Product')
# json_utils.write_json_to_file('./out/rs_Product.json', rs)
# print('done.')
```

```python
# Survey
total=oc.delegator.findCountByCondition('Survey', None,None,None)
print('total records for Survey:', total)

rs=proc_entity_recs('Survey')
# json_utils.write_json_to_file('./out/rs_Survey.json', rs)
json_utils.write_json_to_file('./data/ofbiz/rs_Survey.json', rs)
print('done.')
```
