```python
import json_utils
import sagas.graph.graph_manager as g
from sagas.ofbiz.entity_graph import EntityGraph

reset=True
schema_head="Survey"

eg=EntityGraph()
# gm = g.GraphManager()
if reset:
    eg.setup_schema_head(schema_head)
```

```python
file_name='./out/rs_Survey.json'
ps = json_utils.read_json_file(file_name)
eg.gm.add_object(ps)
```

```python
query = """{
  find_survey(func: eq(surveyId, "1000")) {
    uid
    mo_type
    surveyId
    description
    createdStamp
  }
}
"""
rs=eg.gm.query(query)
print(json.dumps(rs, indent=True))
```

```python
query = """{
  find_survey(func: has(description)) @filter(eq(surveyId, "1000")) {
    uid
    mo_type
    surveyId
    description
    createdStamp
  }
}
"""
rs=eg.gm.query(query)
print(json.dumps(rs, indent=True))
```
